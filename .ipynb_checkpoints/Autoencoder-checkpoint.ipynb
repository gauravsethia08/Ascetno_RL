## Util Functions

```python
def show(objs, titles=None, structure='LR', figsize=(5, 5), clims=False, axes=False, xlabels=False, title_colors=False):
    '''
    utility function to show plots (subplots)
    '''
    plt.style.use('default')
    plt.figure(figsize=figsize)
    for i, obj in enumerate(objs):
        if structure == 'LR': 
            plt.subplot(1, len(objs), i+1)
        elif structure == 'TB': 
            plt.subplot(len(objs), 1, i+1)
        plt.imshow(obj)
        
        if axes: 
            plt.axis(axes[i])
        else: 
            plt.axis('off')
        
        if xlabels: 
            plt.xlabel(xlabels[i])
        
        if clims: 
            plt.clim(clims[i])
        
        if titles:
            plt.title(titles[i], color=title_colors[i] if title_colors else 'teal')
    plt.show()
```

## Generators

```python
def custom_generator(df,
    image_dim=(448, 448),
    batch_size=4,
    seed=0
):
    '''
    Datagen from Dataframe
     * coloums must contain: img_path & mask_path
    
    Yeilds:
        batch of images, batch of masks
    '''
    
    datagen_image = tf.keras.preprocessing.image.ImageDataGenerator(
        rescale=1./255,
#         brightness_range = [0.7, 1.3]
    )
    generator_image = datagen_image.flow_from_dataframe(
        df,
        x_col='img_path',
        target_size=image_dim,
        color_mode='grayscale',
        class_mode="input",
        batch_size=batch_size,
        seed=seed
    )
    
#     datagen_mask = tf.keras.preprocessing.image.ImageDataGenerator(
# #         rescale=1./255,
# #         brightness_range = [0.7, 1.3]
#     )
#     generator_mask = datagen_mask.flow_from_dataframe(
#         df,
#         x_col='out_img_path',
#         target_size=image_dim,
#         color_mode='rgb',
#         class_mode=None,
#         batch_size=batch_size,
#         seed=seed
#     )
    
    while True:
        batch_of_images, batch_of_output  = generator_image.next()
#         batch_of_output = generator_mask.next()
#         print(np.unique(batch_of_output), np.floor(np.sum(batch_of_output)/(255)))
        yield batch_of_images, batch_of_output
```

```python
def check_generator(generator, suptitle):
    '''
    Check the generator
    '''
    batch_of_images, batch_for_decoder =  next(generator)
    x = batch_for_decoder
    
    x_image, y_image = [
        np.squeeze(batch_of_images),
        np.squeeze(batch_for_decoder)
    ]
    
    show(
        [x_image, y_image],
        ['Input Image', "Output Image"],
        clims=[[0, 1], [0, 1]],
        axes=['image', 'image'],
        figsize=((10, 7))
    )

check_generator(custom_generator(main_df[:250], batch_size=2, seed=np.random.randint(10, 155)), 'Check Generator')
```

## Building Model Architecture

```python
def build_model(input_shape=(448, 448, 1), summary=True):
    
    # Input Layer
    input_layer = tf.keras.layers.Input(shape = input_shape)
    
    # ---------------------------------------------------------------------------
    # Encoding Blocks
    # --------------------------------------------------------------------------- 
    conv_1 = tf.keras.layers.Conv2D(32, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(input_layer)
#     conv_2 = tf.keras.layers.Conv2D(32, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(conv_1)
    pool_1 = tf.keras.layers.MaxPooling2D((2,2), strides = (2,2))(conv_1)
    
    conv_3 = tf.keras.layers.Conv2D(64, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(pool_1)
#     conv_4 = tf.keras.layers.Conv2D(64, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(conv_3)
    pool_2 = tf.keras.layers.MaxPooling2D((2,2), strides = (2,2))(conv_3)
    
    conv_5 = tf.keras.layers.Conv2D(128, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(pool_2)
    pool_3 = tf.keras.layers.MaxPooling2D((2,2), strides = (2,2))(conv_5)
    
    conv_6 = tf.keras.layers.Conv2D(128, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(pool_3)
    pool_4 = tf.keras.layers.MaxPooling2D((2,2), strides = (2,2))(conv_6)
    
    # ---------------------------------------------------------------------------
    # Bottle Neck Dense Blocks
    # ---------------------------------------------------------------------------
    # Encoding Bottle neck
    dense_1 = tf.keras.layers.Flatten()(pool_4)
#     dense_2 = tf.keras.layers.Dense(1024, activation = "relu", kernel_initializer = "he_uniform")(dense_1)
    
    # Actual Bottlen neck
    dense_3 = tf.keras.layers.Dense(8, activation = "relu", kernel_initializer = "he_uniform")(dense_1)
    
    # Decoding bottle neck
#     dense_4 = tf.keras.layers.Dense(1024, activation = "relu", kernel_initializer = "he_uniform")(dense_3)
    dense_5 = tf.keras.layers.Dense(25088, activation = "relu", kernel_initializer = "he_uniform")(dense_3)
    reshaped = tf.keras.layers.Reshape((14, 14, 128))(dense_5)
    # ---------------------------------------------------------------------------
    # Decoding Blocks
    # ---------------------------------------------------------------------------
    # Adding output of transition 4 to output of transition 3
    deconv_1 = tf.keras.layers.Conv2DTranspose(128, (3,3), strides=(2, 2), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(reshaped)
    deconv_2 = tf.keras.layers.Conv2DTranspose(64, (3,3), strides=(2, 2), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(deconv_1)
    deconv_3 = tf.keras.layers.Conv2DTranspose(64, (3,3), strides=(2, 2), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(deconv_2)
    deconv_4 = tf.keras.layers.Conv2DTranspose(32, (3,3), strides=(2, 2), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(deconv_3)
    deconv_5 = tf.keras.layers.Conv2DTranspose(32, (3,3), strides=(2, 2), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(deconv_4)
    output = tf.keras.layers.Conv2D(3, (3,3), padding = "same", activation = "relu", kernel_initializer = "he_uniform")(deconv_5)
    
    # Putting everything together and building the model
    model = tf.keras.models.Model(input_layer, output)
    
    print('* Model successfully built.. with {} parameters'.format(model.count_params()))
    if summary: model.summary()
    return model


def compile_model(model):
    '''
    Compile the Model with custom losses
    '''
    model.compile(
        optimizer = tf.keras.optimizers.SGD(),
        loss = tf.keras.losses.MeanAbsoluteError()
    )
    print('* Model successfully compiled..')
    return model
```

```python
model = build_model(summary = True)
model = compile_model(model)
```
